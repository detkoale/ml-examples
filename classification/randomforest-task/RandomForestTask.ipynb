```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.cross_validation import KFold
import numpy as np
```

```python
data = pd.read_csv('abalone.csv')
#print data
```

```python
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
```

```python
#print data.columns
```

```python
#columns from 0 to len-1 contains features 
X = data[data.columns[:-1]]
#last column contains target 
y = data[data.columns[-1:]]
```

```python
crossvalidation = KFold(random_state=1, n_folds=5, n=len(X), shuffle=True)
```

```python
#train randon forest model
for i in np.arange(0,50):
    print "Train random forest with estimator=",i
    clf = RandomForestRegressor(n_estimators=i+1, random_state=1)
    for train_index,test_index in crossvalidation:
            clf.fit(X[train_index], y[test_index])
    score = r2_score(y, clf.predict(X))
    print "R2 metric =", score
```
