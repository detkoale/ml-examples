```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Будем работать с распределением Хи квадрат с парамером k = 4
x = np.linspace(0,20,1000)
for k in [4]:
    rv = sts.chi2(k)
    #рисуем гисограмму
    plt.hist(rv.rvs(1000), normed=True)
    pdf = rv.pdf(x)
    #рисуем график плотности
    plt.plot(x, pdf, label="pdf")
plt.legend()
plt.title("PDF ($\chi^2_k$)")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
k = 4

# Берем 3 значение n: 2,10 и 100 
for n in [1,5,100]:
    
    #Создаем случайную величину и массив для средних значений
    rv = sts.chi2(k)
    sred = []
    
    #генерируем 1000 выборок объема n и заполняем массив средними значениями с каждой выборки
    for z in range(1000):
        a = rv.rvs(n)
        sred.append(sum(a)/n)
    
    #создаем новый график и рисуем гистограмму распределения выборочных средних
    fig = plt.figure()
    plt.hist(sred, bins = 20, normed=True)
    plt.title("n=%s" % n)
    
    #Считаем значение параметров нормального распределениея 
    #Мат ожидание = k, Дисперсия = 2k/n, (для получения сигмы - берем корень из дисперсии)
    ex = k
    dx = mt.sqrt((2*k+0.0)/n)
    
    #для красоты берем точки из области 4 сигм и строим нормальное распределение с EX и DX 
    x = np.linspace(ex-4*dx,ex+4*dx,50)
    norm_rv = sts.norm(ex, dx)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label="norm(%s, %s)" % (ex, round(dx,2)))
    
    #подписываем оси
    plt.ylabel('Hist of avg (x)$_n$')
    plt.xlabel('$x$')
    plt.legend ()
```
