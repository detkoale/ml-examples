# Распределение хи-квадрат

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm
%matplotlib inline
```

## Выборка объема n=5

Генерируем 1000 выборок размера n=5 из распределения хи-квадрат при k=5:

```python
sample5 = chi2.rvs(df=5,size=5*1000).reshape(5,1000)
```

Матожидание распределения хи-квадрат: $MX=k=5$

Дисперсия распределения хи-квадрат: $DX=2k=10$

Матожидание нормального распрееделения аппроксимирующего выборочные средние: $MX=k=5$

Стандартное отклонение нормального распрееделения аппроксимирующего выборочные средние: $\sqrt{\frac{DX}{n}} = \sqrt{\frac{2k}{n}} = \sqrt{\frac{2*5}{5}} = 1.414$

Расчитываем среднее:

```python
means_vector5=sample5.mean(axis=0)
```

Гистограмма:

```python
plt.hist(means_vector5, bins=15, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')

norm_rv = norm(5,1.414)
x = np.linspace(0,10,1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
```
